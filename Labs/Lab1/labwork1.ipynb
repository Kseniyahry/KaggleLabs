```python
import tensorflow as tf
import numpy as np

num_inputs = 3

def generate_xor_data(n_inputs):
    X = np.array(np.meshgrid(*[[0, 1]]*n_inputs)).T.reshape(-1, n_inputs)
    y = np.mod(np.sum(X, axis=1), 2).reshape(-1, 1)
    return X, y

X, y = generate_xor_data(num_inputs)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(8, input_dim=num_inputs, activation='tanh'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(X, y, epochs=600)
loss, accuracy = model.evaluate(X, y)

loss, accuracy = model.evaluate(X, y)
print(f"Loss: {loss:.4f}, Accuracy: {accuracy:.4f}")

predictions = model.predict(X)
print("Predictions (rounded):")
print(np.round(predictions))
print("Actual outputs:")
print(y)
```
